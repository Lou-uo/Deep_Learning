```python
# 3.1 对象的转置
import pandas as pd

# 创建畸形df
v = [ [53, 64, 72, 82], ['女','男','男','女'] ]
i = ['年龄','性别']
c = ['1号','2号','3号','4号']
df = pd.DataFrame(v, index=i, columns=c)
print(df)

# 转置
df = df.T
print(df)
```

```python
# 3.2 对象的翻转

# 左右翻转
df = df.iloc[ : , : : -1]   # 全部的行，全部的列，采样间隔为-1（即倒着采样
print(df)

# 上下翻转
df = df.iloc[ : : -1, : ]
print(df)
```

```python
# 3.3 对象的重塑

import pandas as pd

# 数组法创建sr
i = ['1号', '2号', '3号', '4号']
v1 = [ 10, 20, 30, 40 ]
v2 = [ '女', '男', '男', '女']
v3 = [1,2,3,4]
sr1 = pd.Series(v1, index=i)
sr2 = pd.Series(v2, index=i)
sr3 = pd.Series(v3, index=i)
print(sr1)
print(sr2)
print(sr3, '\n')

# 字典法创建df
df = pd.DataFrame( {'年龄':sr1, '性别':sr2} )
print(df, '\n')

# 把sr并入df中
df['牌照'] = sr3
print(df, '\n')

# 把df['年龄']分离成sr4
sr4 = df['年龄']
print(sr4, '\n')
```

```python
# 3.4 对象的拼接

import pandas as pd

## 一维对象的合并
# 创建sr1和sr2
v1 = [10, 20, 30, 40]
v2 = [40, 50, 60]
k1 = ['1号', '2号', '3号', '4号']
k2 = ['4号', '5号', '6号']
sr1 = pd.Series(v1, index=k1)
sr2 = pd.Series(v2, index=k2)
print(sr1)
print(sr2, '\n')

# 合并
sr3 = pd.concat([sr1, sr2], axis=0)
print(sr3, '\n')
# axis=0--列方向合并; axis=1--行方向合并

# 查看属性
print(sr3.index)
print(sr3.values, '\n')

# 查重
print(sr3.index.is_unique, '\n')  # 有重复返回False


## 一维与二维的合并
v1 = [10, 20, 30]
v2 = ['女', '男', '男']
sr1 = pd.Series(v1, index=['1号', '2号', '3号'])
sr2 = pd.Series(v2, index=['1号', '2号', '3号'])
print(sr1)
print(sr2, '\n')

# 创建df
df = pd.DataFrame( {'年龄':sr1, '性别':sr2} )
print(df, '\n')

# 加上一列特征
df['牌照'] = [1, 2, 3]
print(df, '\n')

# 加上一行个体
df.loc['4号'] = [40, '女', 4]
print(df, '\n')


## 二维对象的合并
v1 = [[10,'女'],[20,'男'],[30,'男'],[40,'女']]
v2 = [[1,'是'],[2,'是'],[3,'是'],[4,'否']]
v3 = [[50,'男',5,'是'],[60,'女',6,'是']]
i1 = ['1号','2号','3号','4号']
i2 = ['1号','2号','3号','4号']
i3 = ['5号','6号']
c1 = ['年龄','性别']
c2 = ['牌照','ikun']
c3 = ['年龄','性别','牌照','ikun']
df1 = pd.DataFrame(v1, index=i1, columns=c1)
df2 = pd.DataFrame(v2, index=i2, columns=c2)
df3 = pd.DataFrame(v3, index=i3, columns=c3)
print(df1)
print(df2)
print(df3)
```
