```python
# 5.1 发现缺失值
import pandas as pd

# 一维
v = [53, None, 72, 82]
k = ['1号', '2号', '3号', '4号']
sr = pd.Series(v,index=k)
print(sr)

# 发现sr的缺失值
print(sr.isnull(), '\n')

# 二维
v = [[None,1],[64,None],[72,3],[82,4]]
i = ['1号','2号','3号','4号']
c = ['年龄','牌照']
df = pd.DataFrame(v,index=i,columns=c)
print(df)

# 发现df的缺失值
print(df.isnull())
print(~df.isnull(), '\n')
```

```python
# 5.2 剔除缺失值
import pandas as pd

# 一维
v = [53, None, 72, 82]
k = ['1号', '2号', '3号', '4号']
sr = pd.Series(v,index=k)
print(sr)

# 剔除sr的缺失值
print(sr.dropna(), '\n')

# 二维
v = [[None,None],[64,None],[72,3],[82,4]]
i = ['1号','2号','3号','4号']
c = ['年龄','牌照']
df = pd.DataFrame(v,index=i,columns=c)
print(df)

# 剔除df含缺失值的行
print(df.dropna(), '\n')
# 剔除df含缺失值的列(非常不推荐这样操作！！！)
print(df.dropna(axis=1), '\n')

# 剔除df全是NaN的个体
print(df.dropna(how='all'))
```

```python
# 5.3 填充缺失值
import pandas as pd
import numpy as np

## 一维
v = [53, None, 72, 82]
k = ['1号', '2号', '3号', '4号']
sr = pd.Series(v,index=k)
print(sr)

# 用常数(0)填充
print(sr.fillna(0))

# 用常数(均值)填充
print(sr.fillna(np.mean(sr)))

# 用前值填充
print(sr.ffill())

# 用后值填充
print(sr.bfill(), '\n')

## 二维
v = [[None,None],[64,None],[72,3],[82,4]]
i = ['1号','2号','3号','4号']
c = ['年龄','牌照']
df = pd.DataFrame(v,index=i,columns=c)
print(df)

# 用常数(0)填充
print(df.fillna(0))

# 用常数(均值)填充
print(df.fillna(np.mean(sr)))

# 用前值填充
print(df.ffill())

# 用后值填充
print(df.bfill(), '\n')
```
