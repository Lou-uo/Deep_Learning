```python
# 7.4 数据透视
import pandas as pd

# 导入Pandas对象
df = pd.read_csv('Titanic_train.csv', index_col=0)
# print(df.head(), '\n')

# 一个特征：性别(pivot_table()中参数aggfunc使用mean方法计算)
print(df.pivot_table('Survived', index='Sex'))

# 两个特征：性别、船舱等级
print(df.pivot_table('Survived', index='Sex', columns='Pclass'), '\n')

## 多个特征
# 重置年龄列
age = pd.cut(df['Age'], [0,25,120])  # 以25岁为分水岭
print(age, '\n')

# 三个特征：性别、船舱等级、年龄
print(df.pivot_table('Survived', index=['Pclass','Sex'], columns=age, observed=True), '\n')
# observed=False 会生成更多组合（包括不存在的），
# observed=True 只生成实际存在的组合。

# 重置费用列
fare = pd.qcut(df['Fare'], 3)    # 自动分为两部分（不推荐）
print(fare, '\n')

# 四个特征：性别、船舱等级、年龄、费用
print(df.pivot_table('Survived', index=['Pclass',fare], columns=['Sex',age], observed=True))
```
