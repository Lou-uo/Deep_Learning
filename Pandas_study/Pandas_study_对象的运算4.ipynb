```python
# 4.1 对象与系数之间的运算

import pandas as pd

# 一维
sr = pd.Series([53, 64, 72], index=['1号', '2号', '3号'])
print(sr)

sr = sr + 10
print(sr)

sr = sr * 10
print(sr)

sr = sr ** 2
print(sr, '\n')

# 二维
v= [[53, '女'], [64, '男'], [72, '男']]
df = pd.DataFrame(v, index=['1号', '2号', '3号'], columns=['年龄', '性别'])
print(df)

df['年龄'] = df['年龄'] + 10
print(df)

df['年龄'] = df['年龄'] * 10
print(df)

df['年龄'] = df['年龄'] ** 2
print(df)
```

```python
# 4.2 对象之间的运算
import pandas as pd
import numpy as np

## 一维
# 创建sr1
v1 = [10, 20, 30, 40]
k1 = ['1号', '2号', '3号', '4号']
sr1 = pd.Series(v1, index=k1)
print(sr1)

# 创建sr2
v2 = [1, 2, 3]
k2 = ['1号', '2号', '3号']
sr2 = pd.Series(v2, index=k2)
print(sr2)

print(sr1 + sr2)
print(sr1 - sr2)
print(sr1 * sr2)
print(sr1 / sr2)
print(sr1 // sr2)
print(sr1 % sr2)
print(sr1 ** sr2)


## 二维
v1 = [[10,'女'],[20,'男'],[30,'男'],[40,'女']]
v2 = [1, 2, 3, 6]
i1 = ['1号','2号','3号','4号']
i2 = ['1号','2号','3号','6号']
c1 = ['年龄','性别']
c2 = ['牌照']
df1 = pd.DataFrame(v1, index=i1, columns=c1)
df2 = pd.DataFrame(v2, index=i2, columns=c2)
print(df1)
print(df2)

df1['加法'] = df1['年龄'] + df2['牌照']
df1['减法'] = df1['年龄'] - df2['牌照']
df1['除法'] = df1['年龄'] / df2['牌照']
df1['乘法'] = df1['年龄'] * df2['牌照']
df1['幂方'] = df1['年龄'] ** df2['牌照']
print(df1, '\n')

print(np.abs(df1['年龄']))
print(np.cos(df1['年龄']))
print(np.exp(df1['年龄']))
print(np.log(df1['年龄']), '\n')

print(df1)
# 布尔型数组(操作与Numpy相同)
print(np.sum(df1['年龄'] > 20))
print(df1['年龄'][df1['年龄'] > 20])
```
